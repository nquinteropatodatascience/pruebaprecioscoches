```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, ensemble
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X, y = data[["curbweight","horsepower","enginesize"]],data.price
#X, y = pd.DataFrame(data.horsepower),pd.DataFrame(data.price).values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=13
)

params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error"
}
```

```python
#https://towardsdatascience.com/gradient-boosting-classification-explained-through-python-60cc980eeb3d
```

```python
reg = ensemble.GradientBoostingRegressor()
reg.fit(X_train, y_train)

mse = mean_squared_error(y_test, reg.predict(X_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```
